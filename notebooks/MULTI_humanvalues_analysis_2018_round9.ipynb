```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
raw = pd.read_csv("../data/MULTI_hv_parties_9_raw.csv")
pd.set_option('display.max_columns', 500)
```

```python
raw.head()
```

```python
cols = {'ipcrtiv':'new_ideas',
        'imprich':'rich',
        'ipeqopt':'equality',
        'ipshabt':'abilities',
        'impsafe':'safety',
        'impdiff':'new_diff',
        'ipfrule':'obey',
        'ipudrst':'understand_diff',
        'ipmodst':'modesty',
        'ipgdtim':'good_time',
        'impfree':'freedom',
        'iphlppl':'help_others',
        'ipsuces':'success',
        'ipstrgv':'strong_gov',
        'ipadvnt':'adventures',
        'ipbhprp':'behave',
        'iprspot':'get_respect',
        'iplylfr':'loyalty',
        'impenv':'nature',
        'imptrad':'traditions',
        'impfun':'fun_pleasure',
        'cntry':'country',
        'essround':'round',
        'lrscale':'leftright_scale',
        'Unnamed: 30':'round'}
```

```python
raw_rename = raw.rename(columns=cols).drop(columns=['idno','leftright_scale','round'])
```

```python
raw_rename.head()
```

### Now we slice our raw dataframe to get individual dfs per country

#### DE

```python
DE_raw = raw_rename[raw_rename["country"] == "DE"]
```

```python
DE_9=DE_raw.drop(columns=['prtvede2',
                          'prtvtdfr',
                          'prtvtfhu',
                          'prtvtcit', 
                          'prtvtdpl',])
```

```python
DE_9 = DE_9.rename(columns={'prtvede1':'party'}).reset_index(drop=True)
```

```python
DE_9[['party']] = DE_9[['party']].astype(int)
```

```python
DE_9.head()
```

```python
DE_9.shape
```

#### In order for our analysis to be more precise, we drop the rows were values for each human value are 7(refusal) or 8(don´t know)

```python
DE_9 = DE_9[(~DE_9["new_ideas"].isin([7,8]))
            &(~DE_9["rich"].isin([7,8]))
            &(~DE_9["equality"].isin([7,8]))
            &(~DE_9["abilities"].isin([7,8]))
            &(~DE_9["safety"].isin([7,8]))
            &(~DE_9["new_diff"].isin([7,8]))
            &(~DE_9["obey"].isin([7,8]))
            &(~DE_9["understand_diff"].isin([7,8]))
            &(~DE_9["modesty"].isin([7,8]))
            &(~DE_9["good_time"].isin([7,8]))
            &(~DE_9["freedom"].isin([7,8]))
            &(~DE_9["help_others"].isin([7,8]))
            &(~DE_9["success"].isin([7,8]))
            &(~DE_9["strong_gov"].isin([7,8]))
            &(~DE_9["adventures"].isin([7,8]))
            &(~DE_9["behave"].isin([7,8]))
            &(~DE_9["get_respect"].isin([7,8]))
            &(~DE_9["loyalty"].isin([7,8]))
            &(~DE_9["nature"].isin([7,8]))
            &(~DE_9["traditions"].isin([7,8]))
            &(~DE_9["fun_pleasure"].isin([7,8]))]
```

```python
DE_9.shape
```

```python
#### We divide our df in 3: 
#### de_ext: those who voted for AfD (party==6)  or NPD(party==8)
#### de_novot: those who didn´t vote (party == 66)
#### de_noext: the rest (also excluding those who didn´t answer(77) or didn´t know(88)
```

```python
DE_ext = DE_9[DE_9["party"].isin([6,8])]
DE_ext.shape
```

```python
DE_novot = DE_9[DE_9["party"] == 66]
DE_novot.shape
```

```python
DE_noext = DE_9[~DE_9["party"].isin([6,8,66,77,88])]
DE_noext.shape
```

```python
DE_ext.describe()
```

```python
DE_noext.describe()
```
