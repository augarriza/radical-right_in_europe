```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
raw = pd.read_csv("../data/MULTI_hv_parties_9_raw.csv")
pd.set_option('display.max_columns', 500)
```

```python
raw.head()
```

```python
cols = {'ipcrtiv':'new_ideas',
        'imprich':'rich',
        'ipeqopt':'equality',
        'ipshabt':'abilities',
        'impsafe':'safety',
        'impdiff':'new_diff',
        'ipfrule':'obey',
        'ipudrst':'understand_diff',
        'ipmodst':'modesty',
        'ipgdtim':'good_time',
        'impfree':'freedom',
        'iphlppl':'help_others',
        'ipsuces':'success',
        'ipstrgv':'strong_gov',
        'ipadvnt':'adventures',
        'ipbhprp':'behave',
        'iprspot':'get_respect',
        'iplylfr':'loyalty',
        'impenv':'nature',
        'imptrad':'traditions',
        'impfun':'fun_pleasure',
        'cntry':'country',
        'essround':'round',
        'lrscale':'leftright_scale',
        'Unnamed: 30':'round'}
```

```python
raw_rename = raw.rename(columns=cols)
```

```python
#We fill the NaN values on colum "round" with 9
raw_rename[['round']] = raw_rename[['round']].fillna(value=9)
```

```python
raw_rename.head()
```

## Now we slice our raw dataframe to get individual df per country

## DE

```python
DE_raw = raw_rename[raw_rename["country"] == "DE"]
```

```python
DE_9=DE_raw.drop(columns=['prtvede2',
                          'prtvtdfr',
                          'prtvtfhu',
                          'prtvtcit', 
                          'prtvtdpl',])
```

```python
DE_9 = DE_9.rename(columns={'prtvede1':'party'}).reset_index(drop=True)
```

```python
DE_9[['party','round']] = DE_9[['party','round']].astype(int)
```

```python
DE_9.head()
```

```python
DE_9["party"].value_counts()
```

#### We divide our df in 3: 
#### de_ext: those who voted for AfD (party==6) or NPD (party==8) 
#### de_novot: those who didn´t vote (party == 66)
#### de_noext: the rest (also excluding those who didn´t answer(77) or didn´t know(88)

```python
DE_ext = DE_9[DE_9["party"].isin([6,8])]
```

```python
DE_novot = DE_9[DE_9["party"] == 66]
```

```python
DE_noext = DE_9[~DE_9["party"].isin([6,8,66,77,88])]
```

```python
DE_ext.describe()
```

```python
DE_noext.describe()
```

```python
DE_novot.describe()
```
